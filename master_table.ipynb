Merging Audrey's Merged csv with Yra's Database

```python
import pandas as pd
```

```python
df = pd.read_csv('merged/Merged_Dataset.csv')
df.head()
```

```python
df.columns = [c.lower() for c in df.columns]
df = df.rename(columns={'streetaddress': 'address'})
```

```python
df.head()
```

```python
df['full_address'] = df['address'].str.title().str.strip()
```

```python
df.head()
```

```python
df.to_csv('merged/Merged_Dataset_clean.csv', index=False)
print('Saved, rows:', len(df))
```
